### Ejercicio Memorias 
## Limpieza de datos

Haciendo uso de la data ‘Memorias_2.xlsx’, determinar:

1.	Eliminar: nombre
2.	Poner la primera letra en mayúscula en los encabezados. Y colocar las medidas cuando sean necesarias.
3.	Conteo de nulos.
4.	Precio: Integrar el símbolo de pesos y quitarle los decimales
5.	Cantidad: Sólo numérico
6.	Cantidad: Imputar con el promedio sin decimales
7.	Capacidad: Sólo numérico
8.	Capacidad: Imputar 
9.	Velocidad: Sólo numérico
10.	Velocidad: Imputar
11.	Formato: Imputar
12.	Calificación: Imputar con 1 decimal
13.	Crear una columna donde muestre el precio con descuento del 20%
14.	Porcentaje de valores únicos en: Cantidad, Capacidad y Formato.


```python
import pandas as pd
import numpy as np

df_memorias = pd.read_excel('../documentos/Memorias_2.xlsx')
```

```python
df_memorias
```

```python
# 1. Eliminar: nombre
df_delete_name =  df_memorias.drop(columns=['nombre'])
df_delete_name
```

```python
# 2. Poner la primera letra en mayúscula en los encabezados. Y colocar las medidas cuando sean necesarias.
df_capitalize = df_memorias.rename(
    columns=lambda col: (
        'Capacidad (GB)' if col.lower() == 'capacidad'
        else 'Velocidad (MB/s)' if col.lower() == 'velocidad'
        else col.capitalize()
    )
)
df_capitalize
```

```python
# 3. Conteo de nulos.
df_memorias.isnull().sum()
```

```python
# 4. Precio: Integrar el símbolo de pesos y quitarle los decimales
df_price = df_memorias.copy() 
df_price['precio'] = df_price['precio'].astype(int).astype(str).radd('$')
df_price
```

```python
# 5. Cantidad: Sólo numérico
df_only_num = df_memorias.copy()
df_only_num['cantidad'] = (
    df_only_num['cantidad']
    .str.extract(r'(\d+)')
    .squeeze()
    .pipe(pd.to_numeric, errors='coerce')
    .fillna(0)
    .astype(int)
)
df_only_num
```

```python
# 6. Cantidad: Imputar con el promedio sin decimales
df_only_num_NaN = df_memorias.copy()
df_only_num_NaN['cantidad'] = (
    df_only_num_NaN['cantidad']
    .astype(str)                           
    .str.extract(r'(\d+)')                 
    .squeeze()                             
    .pipe(pd.to_numeric, errors='coerce')  
)
df_only_num_NaN.fillna({'cantidad':df_only_num_NaN['cantidad'].mean().astype(int)}, inplace=True)
df_only_num_NaN.isnull().sum()
```

```python
# 7. Capacidad: Sólo numérico
df_only_num_NaN_cap = df_memorias.copy()
df_only_num_NaN_cap["capacidad"] = (
    df_only_num_NaN_cap["capacidad"]
    .astype(str)                           
    .str.extract(r'(\d+)')                 
    .squeeze()                             
    .pipe(pd.to_numeric, errors="coerce")  
)
df_only_num_NaN_cap
```

```python
# 8. Capacidad: Imputar 
df_only_num_NaN_cap.fillna({'capacidad':df_only_num_NaN_cap['capacidad'].mean().astype(int)}, inplace=True)
df_only_num_NaN_cap.isnull().sum()
```

```python
# 9. Velocidad: Sólo numérico
df_only_num_NaN_vel = df_memorias.copy()
df_only_num_NaN_vel['velocidad'] = (
    df_only_num_NaN_vel['velocidad']
    .astype(str)
    .str.extract(r'(\d+)')
    .squeeze()
    .pipe(pd.to_numeric, errors='coerce')
)
df_only_num_NaN_vel
```

```python
# 10. Velocidad: Imputar
df_only_num_NaN_vel.fillna({'velocidad':df_only_num_NaN_vel['velocidad'].mean().astype(int)}, inplace=True)
df_only_num_NaN_vel
```

```python
# 11. Formato: Imputar
df_imp_formato = df_memorias.copy()
df_imp_formato['formato'] = df_imp_formato['formato'].fillna('SinFormato')
df_imp_formato
```

```python
# 12. Calificación: Imputar con 1 decimal
df_imp_decimal = df_memorias.copy()
df_imp_decimal['calificacion'] = df_imp_decimal['calificacion'].fillna(round(df_imp_decimal['calificacion'].mean(), 1))

df_imp_decimal
```

```python
# 13. Crear una columna donde muestre el precio con descuento del 20%
df_descuento = df_memorias.copy()
df_descuento['descueto (20%)'] = round(df_descuento['precio'] * .8, 2)
df_descuento
```

```python
# 14. Porcentaje de valores únicos en: Cantidad, Capacidad y Formato.
df_memorias['cantidad'].nunique()
```
