```python
# punto 10
# Mostrar los primeros 10 resultados de los autos Kía entre modelos 1995 y 2015, ordenados de manera ascendente por precio.
print('Los primeros 10 resultados de los autos Kía entre 1995 y 2015 ordenados por crecio son: \n')
df_olx_medallo[(df_olx_medallo['marca'] == 'Kia') &
               ((df_olx_medallo['modelo'] >= 1995) &
                (df_olx_medallo['modelo'] <= 2015))].sort_values(['precio'], ascending=True).head(10)
```

```python
# punto 11
# Calcular el promedio de kilometraje (2 decimales) de los autos marca Ford de color blanco.
df_ford_blanco = df_olx_medallo[(df_olx_medallo['marca'] == 'Ford') &
                                (df_olx_medallo['color'] == 'Blanco')]
print(f'El promedio de kilometraje es: {round(df_ford_blanco['kilometraje'].mean(), 2)}')
```

```python
# Punto 12
# Encontrar los 3 colores más presentes en los autos entre modelos 2000 y 2020
df_3colores_moda = df_olx_medallo[(df_olx_medallo['modelo'] >= 2000) &
               (df_olx_medallo['modelo'] <= 2020)].color.value_counts().head(3)
print(f'Los 3 colores más escogidos entre los años 2000 y 2020 son:')
df_3colores_moda
```

```python
# Punto 13
# Encontrar las 5 primeras marcas más presentes en autos de 4 puertas
df_marcas_4puertas = df_olx_medallo[(df_olx_medallo['puertas'] == 4)].marca.value_counts().head(5)
print('Las 5 marcas de carros mas populares de 4 puertas son: ')
df_marcas_4puertas
```

```python
# Punto 14
# Encontrar las 3 primeras lineas más presentes de Renault
# Filtrar solo los Renault
renault_df = df_olx_medallo[df_olx_medallo["marca"] == "Renault"]
top_3_renault = renault_df["linea"].value_counts().head(3)

print("Las 3 líneas más presentes de Renault son:")
print(top_3_renault)
```

```python
# punto 15
# Calcular el porcentaje (2 decimales) de autos de los diferentes tipos de combustible
df_total_combustibles = df_olx_medallo.combustible.value_counts().sum()
df_gasolina = df_olx_medallo.combustible.value_counts()['Gasolina']
df_diesel = df_olx_medallo.combustible.value_counts()['Diesel']
df_gas_gas = df_olx_medallo.combustible.value_counts()['Gasolina y Gas']
df_elctrico = df_olx_medallo.combustible.value_counts()['Eléctrico']

def porcentaje(combustible, total):
    return round((combustible/total) * 100, 2)

print(f'El {porcentaje(df_gasolina, df_total_combustibles)}% de los autos usan gasolina')
print(f'El {porcentaje(df_diesel, df_total_combustibles)}% de los autos usan diesel')
print(f'El {porcentaje(df_gas_gas, df_total_combustibles)}% de los autos usan gasolina y gas')
print(f'El {porcentaje(df_elctrico, df_total_combustibles)}% de los autos es eléctrico')
```

```python
# punto 15 2.0
# Calcular el porcentaje (2 decimales) de autos de los diferentes tipos de combustible

counts = df_olx_medallo.combustible.value_counts()
total = counts.sum()

def porcentaje(valor, total):
    return round((valor / total) * 100, 2)

for combustible, cantidad in counts.items():
    print(f'El {porcentaje(cantidad, total)}% de los autos usan {combustible.lower()}')
```
